```python
import pandas as pd
data_frame = pd.read_csv('data.csv')
data_frame.shape
```

```python
data_frame.describe()
```

```python
data_frame.values
```

```python
df1 = pd.DataFrame(data_frame,columns = ['team','average_age', 'wins','losses'])
df1['difference'] = df1['wins'] - df1['losses']
df1.sort_values('difference',ascending=False)
```

```python
import seaborn as sns
sns.set()
graph = sns.scatterplot(x='difference',y='average_age',data = data_frame)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ('index',"$index"),
    ("(difference, average_age)", "(@difference, @average_age)"),
    ("team", '@team')
])


p = figure(title='Baseball',x_axis_label='difference',y_axis_label='average_age',plot_width=700,plot_height=700,tools=[TOOLTIPS])
p.circle('difference','average_age',size=10,source=df1)
show(p)
```
